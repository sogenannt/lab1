<h1>Задача 1. Дан набор из $p$ матриц размерностью $(n, n)$ и $p$ векторов размерностью $(n, 1)$, найти сумму произведений матриц на векторы. Написать тесты для кода

```python
import numpy as np

def sum_prod(X, V):
    '''
    X - матрицы (n, n)
    V - векторы (n, ) - одномерные
    Гарантируется, что len(X) == len(V)
    '''
    result = np.zeros(X[0].shape[0])
    for matrix, vector in zip(X, V):
        result += np.dot(matrix, vector)
    return result


matrix_list = [np.array([[1, 2], [3, 4]]), np.array([[5, 6], [7, 8]])]
vector_list = [np.array([2, 3]), np.array([5, 6])]

print(sum_prod(matrix_list, vector_list))
```

```python
import numpy as np

def binarize(M, threshold=0.5):
    return (M > threshold).astype(int)

matrix1 = np.array([[0.3, 0.4], [0.7, 0.5]])
matrix2 = np.array([[400, 16], [57, 117]])

print(binarize(matrix1))
print(binarize(matrix2, 7))
print(binarize(matrix2, 1717))
```

<h1>Задача 3. Напишите функцию, которая возвращает уникальные элементы из каждой строки матрицы. Напишите такую же функцию, но для столбцов. Напишите тесты для кода

```python
import numpy as np

def unique_rows(mat):
    return [np.unique(row) for row in mat]

def unique_columns(mat):
    return [np.unique(col) for col in mat.T]
    
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[7, 7], [8, 8]])
matrix3 = np.array([[2, 3], [2, 6]]) 

print(unique_rows(matrix1))
print(unique_rows(matrix2))
print(unique_rows(matrix3))
print(unique_columns(matrix1))
print(unique_columns(matrix2))
print(unique_columns(matrix3))
```

<h1>Задача 4. Напишите функцию, которая заполняет матрицу с размерами $(m, n)$ случайными числами, распределенными по нормальному закону. Затем считает мат. ожидание и дисперсию для каждого из столбцов и строк, а также строит для каждой строки и столбца гистограмму значений (использовать функцию hist из модуля matplotlib.plot)

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_random_matrix_stats(m, n):
    matrix = np.random.normal(0, 1, (m, n))
    mean_rows = np.mean(matrix, axis=1)
    mean_columns = np.mean(matrix, axis=0)
    var_rows = np.var(matrix, axis=1)
    var_columns = np.var(matrix, axis=0)
    
    plt.figure(figsize=(10, 8))
    plt.subplot(2, 2, 1)
    plt.hist(mean_columns, bins='auto', alpha=0.7, color='black')
    plt.title('Мат. Ожидание по столбцам')
    plt.subplot(2, 2, 2)
    plt.hist(var_columns, bins='auto', alpha=0.7, color='orange')
    plt.title('Дисперсия по столбцам')
    plt.subplot(2, 2, 3)
    plt.hist(mean_rows, bins='auto', alpha=0.7, color='pink')
    plt.title('Мат. Ожидание по строкам')
    plt.subplot(2, 2, 4)
    plt.hist(var_rows, bins='auto', alpha=0.7, color='blue')
    plt.title('Дисперсия по строкам')
    
    plt.tight_layout()
    plt.show()
    
generate_random_matrix_stats(3, 3)
```

<h1>Задача 5. Напишите функцию, которая заполняет матрицу $(m, n)$ в шахматном порядке заданными числами $a$ и $b$. Напишите тесты для кода

```python
import numpy as np
import matplotlib.pyplot as plt

def chess(m, n, a, b):
    board = np.zeros((m, n))
    for i in range(m):
        for j in range(n):
            board[i, j] = a if (i + j) % 2 == 0 else b
    return board

print(chess(4, 4, 7, 0))
print(chess(9, 9, 14, 66))
```

<h1>Задача 6. Напишите функцию, которая отрисовывает прямоугольник с заданными размерами (a, b) на изображении размера (m, n), цвет фона задайте в схеме RGB, как и цвет прямоугольника. Цвета также должны быть параметрами функции. Напишите аналогичную функцию но для овала с полуосями a и b. Напишите тесты для кода.
Примечание: уравнение эллипса (границы овала) можно записать как:
<h1>$\frac{(x-x_0)^2}{a^2}+\frac{(y-y_0)^2}{b^2}=1$

```python
import numpy as np
import matplotlib.pyplot as plt

def draw_picture(picture):
    plt.imshow(picture)
    plt.show()

def draw_rectangle(a, b, m, n, rectangle_color, background_color):
    pic = np.zeros((m, n, 3), dtype=np.uint8)
    pic[:] = background_color
    pic[:b, :a] = rectangle_color
    draw_picture(pic)

def draw_ellipse(a, b, m, n, ellipse_color, background_color):
    pic = np.zeros((m, n, 3), dtype=np.uint8)
    pic[:] = background_color
    y_center, x_center = n // 2, m // 2
    for x in range(m):
        for y in range(n):
            if ((y - y_center) ** 2) / (a ** 2) + ((x - x_center) ** 2) / (b ** 2) <= 1:
                pic[x, y] = ellipse_color
    draw_picture(pic)

draw_rectangle(150, 75, 200, 300, (100, 100, 4), (13, 15, 66))
draw_ellipse(80, 40, 200, 300, (255, 0, 0), (0, 0, 0))
```

<h1>Задача 7. Дан некий временной ряд. Для данного ряда нужно найти его: математическое ожидание, дисперсию, СКО, найти все локальные максимумы и минимумы (локальный максимум - это точка, которая больше своих соседних точек, а локальный минимум - это точка, которая меньше своих соседей), а также вычислить для данного ряда другой ряд, получаемый методом скользящего среднего с размером окна $p$.
<h1>Примечание: метод скользящего среднего подразумевает нахождение среднего из подмножетсва ряда размером $p$

```python
import numpy as np
import matplotlib.pyplot as plt

def find_extremes(timeline, find_max=True):
    max_indices, min_indices = [], []
    for i in range(1, len(timeline) - 1):
        if timeline[i] > timeline[i - 1] and timeline[i] > timeline[i + 1]:
            max_indices.append(i)
        elif timeline[i] < timeline[i - 1] and timeline[i] < timeline[i + 1]:
            min_indices.append(i)
    return max_indices if find_max else min_indices

def rolling_average(timeline, window_size):
    return [float(np.mean(timeline[i:i + window_size])) for i in range(len(timeline) - window_size)]

def analyze_timeline(timeline, window_size):
    timeline = np.array(timeline)
    mean_val = np.mean(timeline)
    variance_val = np.var(timeline)
    stddev_val = np.std(timeline)
    max_indices = find_extremes(timeline)
    min_indices = find_extremes(timeline, find_max=False)
    rolling_avg = rolling_average(timeline, window_size)
    
    print(f"Среднее значение: {mean_val}")
    print(f"Дисперсия: {variance_val}")
    print(f"Стандартное отклонение: {stddev_val}")
    print(f"Индексы максимумов: {max_indices}")
    print(f"Индексы минимумов: {min_indices}")
    print(f"Скользящее среднее: {rolling_avg}")

test_timeline = [2, 4, 3, 6, 8, 7, 5, 1, 9, 12, 11, 14, 13, 10, 8, 6, 5]
analyze_timeline(test_timeline, 4)

```
